Limitation of K-Means Clustering :
1. Specify the cluster number
2. Outliers Sensitive - centroid based
3. Works well only with spherical data

DBSCAN clustering ---> Density Based Special Clustering of Application with Noise
1. No need to specify no. of clusters
2. Density based cluster
3. Robust to outliers
4. Works well with spherical and non spherical data

Features of DBSCAN 
1. Hyperperameter epsilon (area) , Minimum Points (number of entity
2. Sparse means if there is no enough points in its area 


Types of points:
1. Core points has the capacity to make a cluster and expand the cluster
2. Border Points
3. Noise Points

Working Procedure of DBSCAN 
1. choose Epsilon and Min points
2. Identify core points, border points and noise points
3. Clustering the core points based on epsilon neighborhood consideratioon
4. Clustering border points

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import StandardScaler 

from sklearn.datasets import make_moons 
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
```

```python
X, y = make_moons(n_samples=200, noise=0.06, random_state=42) 

X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
scaler = StandardScaler() 

X_scaled = scaler.fit_transform(X) 

plt.scatter(X_scaled[:,0],X_scaled[:,1])
```

```python
dbscan = DBSCAN(eps =0.4,min_samples=3) 
```

```python
dbscan.fit(X_scaled) 

label_dbscan = dbscan.labels_
print(label_dbscan)
```

```python
sns.scatterplot(x = X_scaled[:,0], y = X_scaled[:,1] , hue=label_dbscan, palette='viridis')
```

```python
kmeans = KMeans(n_clusters=2,init='k-means++',max_iter=400,random_state=42) 
kmeans.fit(X_scaled) 
label_kmeans = kmeans.labels_ 
sns.scatterplot(x = X_scaled[:,0], y = X_scaled[:,1] , hue=label_kmeans, palette='viridis')
```
