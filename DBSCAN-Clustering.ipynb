Limitation of K-Means Clustering :
1. Specify the cluster number
2. Outliers Sensitive - centroid based
3. Works well only with spherical data

DBSCAN clustering ---> Density Based Special Clustering of Application with Noise
1. No need to specify no. of clusters
2. Density based cluster
3. Robust to outliers
4. Works well with spherical and non spherical data

Features of DBSCAN 
1. Hyperperameter epsilon (area) , Minimum Points (number of entity
2. Sparse means if there is no enough points in its area 


Types of points:
1. Core points has the capacity to make a cluster and expand the cluster
2. Border Points
3. Noise Points

Working Procedure of DBSCAN 
1. choose Epsilon and Min points
2. Identify core points, border points and noise points
3. Clustering the core points based on epsilon neighborhood consideratioon
4. Clustering border points
