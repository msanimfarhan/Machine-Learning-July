How to select features (Basic way):
1. keep features which has larger variance because it has more effect on the actual output or labels

Working of PCA - 
1. Makes features which hold most variance (principal components n)
2. Makes n numbers principal components combining correlated features (ensuring datas essence)
3. Tranforms existing data based on principal component

It doesnot work when-
1. Correlation is less between data
2. Delete features which has effect on prediction , so we get less accuracy
